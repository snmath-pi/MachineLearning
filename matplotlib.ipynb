```python
#BAR CHATS AND ANALYSING DATA FROM CSV
import numpy as np
import csv
from matplotlib import pyplot as plt
from collections import Counter

with open('/kaggle/input/most-common-programming-languages-used-worldwide/data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))


languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()
plt.barh(languages, popularity)
plt.legend()

plt.title("Most Popular Languages")
plt.xlabel("No of people")

plt.tight_layout()
plt.rcdefaults()
plt.show()
```

```python
#using pandas
#BAR CHATS AND ANALYSING DATA FROM CSV
import pandas as pd
from matplotlib import pyplot as plt
from collections import Counter

data = pd.read_csv('/kaggle/input/most-common-programming-languages-used-worldwide/data.csv')
ids = data['Responder_id']
lang_responses = data['LanguagesWorkedWith']
language_counter = Counter()
    
for row in lang_responses:
    language_counter.update(row.split(';'))
    


languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()
plt.barh(languages, popularity)
plt.legend()

plt.title("Most Popular Languages")
plt.xlabel("No of people")

plt.tight_layout()
plt.rcdefaults()
plt.show()
```
